```python
import pandas as pd
import os
import cv2
```

```python
data = pd.read_csv('../raw/train.csv')

data_0 = data[data["level"] == 0].head(3000)
data_2 = data[data["level"] == 2].head(3000)

data_balance = pd.concat(
    [data_0, data_2]
)
data_balance = data_balance.sample(frac = 1)
data_balance = data_balance.loc[:, ~data_balance.columns.str.contains('Unnamed')]
data_balance = data_balance.reset_index()
display(data_balance)
```

```python
train = data_balance.head(4000)
train = train.loc[:, ~train.columns.str.contains('index')]
data_balance = data_balance.drop(train.index)

val = data_balance.head(1000)
val = val.loc[:, ~val.columns.str.contains('index')]
data_balance = data_balance.drop(val.index)

test = data_balance
test = test.loc[:, ~test.columns.str.contains('index')]

test = test.reset_index()
train = train.reset_index()
val = val.reset_index()
test = test.loc[:, ~test.columns.str.contains('index')]
val = val.loc[:, ~val.columns.str.contains('index')]
train = train.loc[:, ~train.columns.str.contains('index')]

display(train.head())
display(len(val))
display(test.head())
```

```python
train.to_csv("../dataset/balance/train.csv", index=None)
test.to_csv("../dataset/balance/test.csv", index=None)
val.to_csv("../dataset/balance/val.csv", index=None)
```
