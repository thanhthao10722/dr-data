```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = ["id_code", "level"]

data_train = pd.read_csv("round_2/imbalance/train.csv")
data_test = pd.read_csv("round_2/imbalance/test.csv")
```

```python
labels = ['Train', 'Test']
sizes = [len(data_train), len(data_test)]
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=False)
ax1.axis('equal')
plt.show()
fig1.savefig('myimage.png', format='png', dpi=1200)
```
